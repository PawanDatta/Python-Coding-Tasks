## <font color = SlateBlue> Task 2: Data Analysis  </font>

### <font color = MediumSeagreen> Subtask 2.1: Reduce those Digits! </font>

```python
# converting`$` to `million $` (use of budget and gross columns)

movies.budget = movies.budget.apply(lambda x: round(x/1000000, 2))  # Because numbers are in millions, I can afford to round them off to two decimals
movies.Gross = movies.Gross.apply(lambda x: round(x/1000000, 2))    # Because numbers are in millions, I can afford to round them off to two decimals
movies.head()
```

### <font color = MediumSeagreen> Subtask 2.2: Let's Talk Profit! </font>

```python
# Creating a new column named as 'profit' by subtracting the 'budget' column from the 'gross' column

movies['profit'] = movies['Gross'] - movies['budget']
movies.head()
```

```python
# Sorting the dataframe with the 'profit' column as reference

movies.sort_values('profit', ascending=False, inplace=True)
movies.head()
```

```python
# Getting the top 10 profitable movies using position based indexing

top_10_movies = movies.iloc[0:10]
top_10_movies
```

```python
#Preparing for plotting profit vs budget
sns.set_style("dark")
plt.figure(figsize=(8,6))
plt.title("Profit Vs Budget", fontdict={'color': 'blue', 'size':18})
plt.xlabel('Budget (Millions)', fontdict={'color':'brown', 'size':15})
plt.ylabel('Profit (Millions)', fontdict={'color':'brown', 'size':15})


#plotting profit vs budget
plt.scatter(movies.budget, movies.profit)
plt.show()
```

#### <font color = MediumSeagreen> Observation </font>
The dataset consists of 100 best performing movies from the year 2010 to 2016. And the scatter plot relieves following insights:

    1.There are movies with negative profit.

## <font color = SlateBlue> Task 3 : Demographic analysis </font>

### <font color = MediumSeagreen> Subtask 3.1 Combine the Dataframe by Genres </font>

```python
# Creating the dataframe df_by_genre

df_by_genre = pd.concat( (movies.loc[:, ['genre_1', 'genre_2', 'genre_3']], movies.iloc[:, 16:-4]), axis=1)

#sanity check 
df_by_genre.shape
```

```python
# Creating a column cnt and initializing it to 1

df_by_genre['cnt'] = 1

#sanity check 
df_by_genre.head()
```

```python
# Grouping the movies by genres, 

df_by_g1 = df_by_genre.groupby('genre_1').sum()
df_by_g2 = df_by_genre.groupby('genre_2').sum()
df_by_g3 = df_by_genre.groupby('genre_3').sum()
```

```python
# Adding the grouped data frames and storing it in a new data frame

df_add = df_by_g1.add(df_by_g2, fill_value=0).add(df_by_g3, fill_value=0)

#sanity check 
# df_add.head()
```

```python
# Extracting genres with atleast 10 occurences

genre_top10 = df_add[df_add['cnt'] >= 10]

#sanity check 
# genre_top10
```

```python
# Taking the mean for every column by dividing with cnt 

genre_top10.iloc[:, 0:44] = genre_top10.iloc[:, 0:44].div(genre_top10['cnt'], axis=0)

#sanity check 
# genre_top10
```

```python
# Rounding off the columns of Votes to two decimals

genre_top10.loc[:, 'VotesM':'VotesnUS'] = genre_top10.loc[:, 'VotesM':'VotesnUS'].round(2)

#sanity check 
# genre_top10
```

```python
# Converting CVotes to int type

genre_top10[genre_top10.loc[:, 'CVotes10':'CVotesnUS'].columns] = genre_top10[genre_top10.loc[:, 'CVotes10':'CVotesnUS'].columns].astype('int')
genre_top10
```

### <font color = MediumSeagreen> Subtask 3.4: US vs non-US Cross Analysis </font>

**`Inferences:`** Write your two inferences/observations below:

- Inference 1: USA people like USA movies more than non-USA movies. 

- Inference 2: Non-USA people like both kinds of movies fairly equally.
